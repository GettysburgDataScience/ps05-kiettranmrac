### Problem 5

Now that we've inspected the data, let's build a classifier (you'll compare two models) that predicts which passengers survived. These are the steps you should take:

 - Choose your features. The label will be survivorship.
 - Split the data into training and test sets.
 - Standardize the values using StandardScaler
 - Choose two of the following model types: KNN, SVM, Decision Tree. Create those models and select hyper-parameters.
 - Fit your models to the training data.
 - Apply the fit models to create predictions from the test data.
 - For each model, create a confusion matrix for the training and test data. You may want to revisit the hyper-parameters and re-fit the models.

 - Answer the questions at the bottom of the code block.

````python
'''Your code here'''
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import ConfusionMatrixDisplay

# 1: Choose features
features = ['Pclass', 'Age', 'Sex', 'SibSp', 'Parch', 'Fare']
label = ['Survived']

# 2: Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(titanic_df[features], titanic_df[label], test_size = 0.2)

# 3: Standardize the values using StandardScaler

ss = StandardScaler()
ss.fit(X_train)

Xs_train = ss.transform(X_train)
Xs_test = ss.transform(X_test)

ys_train = y_train.values.ravel()
ys_test =  y_test.values.ravel()

# display(Xs_train.shape)

# 4: Choose two of the following model types: KNN, SVM, Decision Tree. Create those models and select hyper-parameters.
knn_clf = KNeighborsClassifier(n_neighbors=26) #sqr(712) ~= 26
svm_clf = SVC(kernel = 'rbf', gamma=1)

# 5: Fit your models to the training data.
knn_clf.fit(Xs_train, ys_train)
svm_clf.fit(Xs_train, ys_train)

# 6: Apply the fit models to create predictions from the test data.
y_knn_train = knn_clf.predict(Xs_train)
y_knn_test = knn_clf.predict(Xs_test)
y_svm_train = svm_clf.predict(Xs_train)
y_svm_test = svm_clf.predict(Xs_test)

# 7: For each model, create a confusion matrix for the training and test data. You may want to revisit the hyper-parameters and re-fit the models.
fig, ax = plt.subplots(2, 2, figsize = (10, 8))

ConfusionMatrixDisplay.from_predictions(ys_train, y_knn_train,
                                        normalize = 'true',
                                        display_labels = ['Did Not', 'Survived'],
                                        ax = ax[0, 0])
ConfusionMatrixDisplay.from_predictions(ys_test, y_knn_test,
                                        normalize = 'true',
                                        display_labels = ['Did Not', 'Survived'],
                                        ax = ax[0, 1])
ConfusionMatrixDisplay.from_predictions(ys_train, y_svm_train,
                                        normalize = 'true',
                                        display_labels = ['Did Not', 'Survived'],
                                        ax = ax[1, 0])
ConfusionMatrixDisplay.from_predictions(ys_test, y_svm_test,
                                        normalize = 'true',
                                        display_labels = ['Did Not', 'Survived'],
                                        ax = ax[1, 1])
ax[0, 0].set_title('Training KNN')
ax[0, 1].set_title('Test KNN')
ax[1, 0].set_title('Training SVM')
ax[1, 1].set_title('Test SVM')
fig.tight_layout()
plt.show()
````
